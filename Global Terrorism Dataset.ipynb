# **Project Name**    - Global Terrorism Dataset



The Global Terrorism Database (GTD) is an open-source database including information on terrorist attacks around the world from 1970 through 2017. The GTD includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period and now includes more than 180,000 attacks. The database is maintained by researchers at the National Consortium for the Study of Terrorism and Responses to Terrorism (START), headquartered at the University of Maryland.

#### Chart - 1  . Number of Terrorist Attacks per Year

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset with the correct encoding
data = pd.read_csv('/content/drive/MyDrive/Global Terrorism Data.csv', encoding='ISO-8859-1', low_memory=False)

# Ensure data is a DataFrame
assert isinstance(data, pd.DataFrame), "Data is not a DataFrame!"

# Group by year and count the number of attacks
attacks_per_year = data.groupby('iyear').size()

# Plotting
plt.figure(figsize=(14, 7))
attacks_per_year.plot()
plt.title('Number of Terrorist Attacks per Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

A line chart is an effective way to visualize trends over time. It allows us to quickly grasp how the number of terrorist attacks has changed year by year.

There was a significant surge in the number of attacks starting from the early 2010s, reaching a peak around the mid-2010s.
After the peak, there seems to be a decline in the number of attacks, but the numbers are still relatively high compared to the 1970s-2000s.

#### Chart - 9 Top Ten Rivalries

```python
# Terrorist organisation countrywise.
gname_country=df[df['organisation']!='Unknown'].copy()
group_attacks=gname_country.groupby(['organisation','Country']).agg({'organisation':'count'}).rename(columns={'organisation':'no_of_attacks'}).reset_index().nlargest(10,'no_of_attacks')
country_mode=gname_country.groupby('organisation')['Country'].apply(lambda x: x.mode().loc[0]).reset_index()
top_df = pd.merge(left=country_mode,right=group_attacks,on=['organisation','Country'],how='left').nlargest(10,'no_of_attacks')
top_df.sort_values('no_of_attacks',ascending=True,inplace=True)
```

# **CHALLENGES FACED**

1. Collecting Meaningful Data

Challenge: The sheer volume of data available can be overwhelming, and not all of it is pertinent or of value.
Solutions:

Define Clear Objectives: Before embarking on data collection, it's vital to have a clear idea of the objectives. Knowing what you're trying to achieve or discover can guide the data collection process.

Data Quality over Quantity: Instead of aiming to gather vast amounts of data, focus on the quality and relevance of the data. High-quality, relevant data provides more actionable insights than a large volume of irrelevant data.

Use Targeted Data Collection Methods: Employ methods that are tailored to gather the specific type of data you need. This could be through surveys, interviews, or specialized sensors, depending on the context.

2. Unknown Values

Challenge: Incomplete datasets with missing or unknown values can skew analysis and lead to inaccurate conclusions.
Solutions:

Imputation: This is a method where missing values are filled in based on the known values of other data. There are various imputation methods, such as mean imputation, regression imputation, or even more advanced techniques like KNN imputation.

Data Cleaning: Before starting the analysis, always perform a thorough data cleaning process. This involves identifying and handling outliers, errors, and missing values.

Acknowledge Limitations: In some cases, it might be best to acknowledge the limitations of the dataset rather than trying to fill in missing values. This ensures that any conclusions drawn are based on authentic, known data.

3. Visual Representation of Data

Challenge: Raw data can be hard to interpret, and creating visualizations manually can be time-consuming.
Solutions:

Use Data Visualization Tools: Tools like Tableau, Power BI, and Python libraries like Matplotlib and Seaborn can automate the process of data visualization.

Templates and Dashboards: Instead of creating visualizations from scratch every time, develop templates or dashboards that can be reused with new data.

Understand Your Audience: The key to a good visualization is understanding who will be viewing it. Tailor your visualizations to your audience's level of expertise and what they need to glean from the data.
